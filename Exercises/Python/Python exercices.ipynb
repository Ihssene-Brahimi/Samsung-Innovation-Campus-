Write a Python function that takes a list and returns a new list with unique elements of the first list.



```

```



````python
#list(set(list))
#a ) set(list)
#a.add(tuple)  okay   a.add(list)  noo
#a.update([list])
mylist = [1,2,3,3,3,3,4,5]
def unique(mylist):
  for x in mylist:
    if mylist.count(x) > 1:
      mylist.remove(x)

  return mylist

unique(mylist)
````

# Exercice 17
Write the definition of a Point class. Objects from this class should have a

a method show to display the coordinates of the point
a method move to change these coordinates.
a method dist that computes the distance between 2 points.
Note the distance between 2 points A(x0, y0) and B(x1, y1) can be compute as follows
$d(A,B)=\sqrt{(x_{1}+x_{0})^{2} + (y_{1}+y_{0})^{2} }$

this is the expected result

```
p1 = Point(2, 3)
>>> p2 = Point(3, 3)
>>> p1.show()
(2, 3)
>>> p2.show()
(3, 3)
>>> p1.move(10, -10)
>>> p1.show()
(12, -7)
>>> p2.show()
(3, 3)
>>> p1.dist(p2)
1.0
```



````python
import math as m 

class Point:
  def __init__(self,x,y):
    self.x = x
    self.y = y

  def show(self):
    return (self.x, self.y)

  def move(self,x,y):
    self.x+=x
    self.y+=y

  def dist(self,p):
    if isinstance(p,Point):
      return m.sqrt(m.pow(p.x-self.x,2)+m.sqrt(m.pow(p.y-self.y,2)))


a = Point(5,5)  
print(a.show())
a.move(2,-6)
print(a.show())
b = Point(0,0)
a.dist(b)
````
